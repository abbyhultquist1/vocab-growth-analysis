```python
import sets
import random

#run numruns times and make new file for each
import sys
k=0
numruns = 100
run = ''

def filename(j):
	j = j+1
	str_list = ['run']
	str_list.append(`j`)
	str_list.append('.txt')
	return ''.join(str_list)
while k < numruns:
	run = filename(k)
	file = open(run, "w")

	month_no = 16
	month_list = '16_semantic0.85.txt'
	allwords = open('30_semantic0.85.txt', 'r')
	combinations = [item.rstrip('\n') for item in allwords]
	combinations = map(lambda x: x.split(","), combinations)
	allwords.close()
	connection_map = {}

	# Map out the connections between words
	for x in combinations:
	   if not x[0] in connection_map:
	       connection_map[x[0]] = []
	   if not x[1] in connection_map:
	       connection_map[x[1]] = []
	   connection_map[x[0]].append(x[1])
	   connection_map[x[1]].append(x[0])

	# Remove duplicates
	for key in connection_map.keys():
	    edgesList = connection_map[key]
	    edgesSet = sets.Set(edgesList)
	    connection_map[key] = edgesSet
	   
	#print connection_map["alligator"]

	# Load up prototypical connections
	knownwords = open(month_list, 'r')
	datalist = knownwords.readlines()
	knownwords.close()
	datalist = [item.rstrip('\n').split(",") for item in datalist]

	# Instantiate primitive network
	import csv
	prototypical_kids_bin = csv.reader(open("prototypical_child_words_binary.csv"))
	transition = month_no - 15
	itsy = 0
	row = prototypical_kids_bin.next()
	while itsy < transition:
		row = prototypical_kids_bin.next()
		itsy = itsy + 1
	#row num is dependent on what mo - mo
	#for prototypical_child_words from MCDI 
	num_desired_tuples = int(row[356])
	print num_desired_tuples
	network = []

	def is_present(target_connection, dataset):
		are_equal_tuples = lambda x, y: x[0] == y[0] and x[1] == y[1]
		for connection in dataset:
			if are_equal_tuples(connection, target_connection):
				return True
		return False

	# Randomly fill network with unique tuples
	i = 0
	while i < num_desired_tuples:
		#choose a random tuple
		random_pair = random.choice(datalist)
		#get the first element of the tuple to have a neighbor added
		lucky_word = random_pair[0]
		#if no possible edges, choose a new tuple
		while (len(connection_map[lucky_word]) == 1):
			random_pair = random.choice(datalist)
			lucky_word = random_pair[0]

		#randomly select a word from possible words to be added as a neighbor
		connection_list = []
		for x in connection_map[lucky_word]:
			connection_list.append(x)
		new_word = random.choice(connection_list)

		if not is_present((lucky_word , new_word), datalist):
			i += 1
			#acquire new_word here for word being added
			#if item in connection_map[new_word] exists 
			#in datalist add edge
			new_word_edgelist = connection_map[new_word]
			
			for connection in datalist:
				if connection[0] in new_word_edgelist and not is_present((new_word, connection[0]), datalist):
					datalist.append((new_word, connection[0]))
					datalist.append((connection[0], new_word))
					#print "Added: ", (new_word, connection[0])


	file.write(str(datalist))
	file.close()
	k = k+1
```
